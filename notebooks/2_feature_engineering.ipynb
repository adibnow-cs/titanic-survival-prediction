```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
train_df = pd.read_csv('../data/train.csv')
print(train_df.head())
```

```python
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
train_df['WomanOrChild'] = ((train_df['Sex'] == 'female') | (train_df['Age'] <= 12)).astype(int)
train_df['IsRich'] = (train_df['Fare'] > 50).astype(int)
train_df['IsFirstClass'] = (train_df['Pclass'] == 1).astype(int)
train_df['HasCabin'] = train_df['Cabin'].notnull().astype(int)
```

```python
train_df.to_csv('../data/train_engineered_1.csv', index=False)
```
